```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def samplePlot():

    x = np.linspace(0, 2 * np.pi, 100)
    y = np.sin(x)

    plt.plot(x, y)
    plt.title("Sine Function")
    plt.xlabel("x (radians)")
    plt.ylabel("sin(x)")
    plt.grid()
    plt.show()
```

```python
samplePlot()
```
